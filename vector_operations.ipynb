```python
plt.quiver(2,2,4,5,scale_units='xy',angles='xy',scale=1)
plt.ylim(-8,8)
plt.xlim(-8,8)
plt.show()
```

```python
plt.quiver(2,2,4,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.ylim(-8,8)
plt.xlim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-4,-5,scale_units='xy',angles='xy',scale=1,color='y')
plt.ylim(-8,8)
plt.xlim(-8,8)
plt.show()
```

```python
#Addition of two vectors
vector_1=np.asarray([0,0,2,3])

vector_2=np.asarray([0,0,3,-2])

sum=vector_1+vector_2
print(sum)
```

```python
#Substraction of two vectors
vector_1=np.asarray([0,0,2,3])

vector_2=np.asarray([0,0,3,-2])

difference=vector_1-vector_2
print(difference)
```

```python
#Multiplication two vectors
vector_1=np.asarray([0,0,2,3])

vector_2=2*vector_1

print(vector_2)
```

```python
#Dot product of two vectors
a=np.array([2,3])

b=np.array([4,4])

a_dot_b=np.dot(a,b)

#2*4+3*4=20
```

```python
print(a_dot_b)
```

```python
#Cross product of two vectors

a=np.array([2,3])

b=np.array([4,4])

a_cross_b=np.cross(a,b)
```

```python
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
boston_dataset=load_boston()
```

```python
boston_df=pd.DataFrame(boston_dataset.data,columns=boston_dataset.feature_names)
```

```python
boston_df.head()
```

```python
#information on the dataframe
boston_df.info()
```

```python
boston_df.mean()
```

```python
boston_df.std()
```

```python
boston_df.min()
```

```python
boston_df.max()
```

```python
boston_df.describe()
```

```python
#adding a column
boston_df["Price"]=boston_dataset.target
```

```python
#removing a row
boston_df.drop(index=0,axis=0)
```

```python
#drop a column
boston_df.drop(columns='ZN',axis=1)
```

```python
#locating row using index value
boston_df.iloc[2]
```

```python
boston_df.shape
```

```python
boston_df.to_csv('boston.csv')
```

```python
#finding the number of missing value
boston_df.isnull().sum()
```
